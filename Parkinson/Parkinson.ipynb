```python
#from parkinson import *

import sys
import os

module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path + "\\common")

from classification import *

scalers = [
    MinMaxScaler((-1, 1))
]

classifiers = [
    xgb.XGBClassifier(),
    RandomForestClassifier()
]


params = [
    {
        #'scaler__feature_range': [(0, 1), (-1, 1)],
        "learning_rate": [0.05, 0.10, 0.15],
        "max_depth": [ 3, 4, 5, 6, 8],
        "min_child_weight": [ 1, 3, 5, 7],
        "gamma":[ 0.0, 0.1, 0.2],
        'eval_metric': ['mlogloss']
    },
    {
        'bootstrap': [True, False],
         'max_depth': [10, 20, 50, 100],
         'max_features': ['auto', 'sqrt'],
         'min_samples_leaf': [1, 2, 4],
         'min_samples_split': [2, 5, 10],
         'n_estimators': [100, 200, 500, 1000,]
    }
]

df = pd.read_csv('parkinsons.data')

detect = Classification('parkinsons.data', 
                         [f for f in df.columns if f != 'status' and f != 'name'],
                         'status'
                        )
for scaler in scalers:
    for clf, params in zip(classifiers, params):
        detect.set_methods(scaler, clf)
        detect.set_parameters(params)
        detect.run()
```
